```python
def order_dataframe_test(df):
    df = df.sort_values('score', ascending=False).reset_index(drop=True)
    index = df.index.to_numpy()
    items_ids = df['item_id'].to_list()
    indexes_locked = [2,3]
    
    mask_locked = np.ones( df.shape[0], dtype=bool)
    
    for i in indexes_locked:
        mask_locked[i] = False
    print(mask_locked)
    vertical_counter = 0
    domain_counter = 0
    last_vertical = None
    last_domain = None
    for i in range(0,len(index)):
        if i not in indexes_locked:
            idx = index[i]
            mask = True
            row = df.loc[idx] 
            if row['vertical'] == last_vertical:
                vertical_counter += 1
            else:
                vertical_counter = 0
            if row['domain'] == last_domain:
                domain_counter += 1
            else:
                domain_counter = 0

            if vertical_counter==1:
                mask = mask & (df.iloc[i::]['vertical'] != last_vertical) & mask_locked[i::]
            if domain_counter==4:
                mask = mask & (df.iloc[i::]['domain'] != last_domain) & mask_locked[i::]

            if isinstance(mask, pd.Series):
                new_idx = df.iloc[i::][mask].iloc[0].name
                index = index[index != new_idx]
                index = np.insert(index,i,new_idx)
                idx = new_idx

            last_vertical = df.loc[idx,'vertical']
            last_domain = df.loc[idx,'domain']
            df = df.reindex(index)
    return df.reindex(index)
```
